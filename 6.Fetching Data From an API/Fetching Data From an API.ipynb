```python
df = pd.DataFrame()
```

```python
df
```

```python
for i in range(1, 513):
    print(f"Fetching page {i}...")
    response = requests.get(
        f'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}'
    )
    data = response.json()
    
    # Check if the API returned 'results'
    if 'results' in data:
        temp_df = pd.DataFrame(data['results'])[
            ['id', 'title', 'overview', 'popularity', 'vote_average', 'vote_count']
        ]
        df = pd.concat([df, temp_df], ignore_index=True)
    else:
        print(f"⚠️ No 'results' on page {i}. Response: {data}")

print("✅ Done!")
```
